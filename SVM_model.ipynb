```python
iris_dataset['flower_name']=iris_dataset.target.apply(lambda x: iris.target_names[x])
iris_dataset
```

```python
iris_dataset_setosa=iris_dataset[iris_dataset['target']==0]
iris_dataset_versicolor=iris_dataset[iris_dataset['target']==1]
iris_dataset_virginica=iris_dataset[iris_dataset['target']==2]
```

```python
plt.scatter(iris_dataset_setosa['sepal length (cm)'],iris_dataset_setosa['sepal width (cm)'],color='red',marker='+')
plt.scatter(iris_dataset_versicolor['sepal length (cm)'],iris_dataset_versicolor['sepal width (cm)'],color='green',marker='.')
```

```python
iris_dataset
```

```python
x=iris_dataset.drop(['target','flower_name'],axis=1)
y=iris_dataset.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=SVC()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
y_test
```

```python
accuracy_score(y_test,predictions)
```
