```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dataset=pd.read_csv("titanic_dataset.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.drop("Cabin",axis=1,inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset
```

```python
dataset.drop(["PassengerId","Name","SibSp","Parch","Ticket"],axis=1,inplace=True)
dataset
```

```python
categorical_embarked=pd.get_dummies(dataset["Embarked"])
categorical_embarked
```

```python
dataset=pd.concat([dataset,categorical_embarked],axis=1)
dataset
```

```python
dataset.drop("Embarked",axis=1,inplace=True)
dataset
```

```python
categorical_sex=pd.get_dummies(dataset["Sex"],drop_first=True)
categorical_sex
```

```python
dataset=pd.concat([dataset,categorical_sex],axis=1)
dataset
```

```python
dataset.drop("Sex",axis=1,inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
x=dataset.drop("Survived",axis=1)
y=dataset["Survived"]
y=y.astype('int')
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
predictions
```

```python
y_test
```

```python
confusion_matrix(y_test,predictions)
```

```python
accuracy_score(y_test,predictions)
```
